```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import warnings
import tslearn as tsl
from tslearn.clustering import TimeSeriesKMeans
```

```python
temp = pd.read_csv('temperature.csv')
temp = temp.dropna().reset_index().drop('index',axis=1)
temp['datetime'] = pd.to_datetime(temp['datetime'])
temp.head()
```

```python
cities_list = temp.columns.tolist()[1:]
cities_list
```

```python
cities_list[4]
```

```python
plt.plot(temp['datetime'],temp[cities_list[4]]);
```

```python
temp.set_index('datetime', inplace=True)
```

```python
td = temp.resample("D").mean()
td.head().round(2)
```

```python
td.Seattle.plot()
```

```python
td.T.head()
```

```python
data_array = td.T.values
data_array.shape
data_array[4]
```

```python
model = TimeSeriesKMeans(n_clusters=3, metric="dtw", max_iter=10)
model.fit(data_array)
```

```python
warnings.filterwarnings("ignore")
y=model.predict(data_array)
```

```python
y
```

```python
x = td.index
```

```python
names = td.T.index
cluster_count = 3
fig, axs = plt.subplots(cluster_count, 1, figsize=(12, 15)) 

for cluster_number in range(cluster_count):
    cluster_data = data_array[y == cluster_number, :]
    cluster_names = names[y == cluster_number]

    for i, time_series in enumerate(cluster_data):
        axs[cluster_number].plot(time_series, alpha=0.5, label=cluster_names[i])
    
    axs[cluster_number].set_title(f'Cluster {cluster_number + 1}')
    axs[cluster_number].legend()

plt.tight_layout()
plt.show()
```
